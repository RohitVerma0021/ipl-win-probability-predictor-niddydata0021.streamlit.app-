```python
deli_df=match_df.merge(deli,on="match_id")
```

```python
deli_df=deli_df[deli_df["inning"]==2]
```

we are calculate the what is the run after bals and it's ball ball record

```python
deli_df["current_score"]=deli_df.groupby('match_id').cumsum()["total_runs_y"]
```

```python
deli_df["runs_left"]=deli_df["total_runs_x"]-deli_df["current_score"]
```

```python
deli_df["batting_team"].unique()
```

```python
deli_df["batting_team"]=deli_df["batting_team"].str.replace('Delhi Daredevils','Dehli Capitals')
deli_df["bowling_team"]=deli_df["bowling_team"].str.replace('Delhi Daredevils','Dehli Capitals')

deli_df["batting_team"]=deli_df["batting_team"].str.replace('Deccan Chargers','Sunrisers Hyderabad')
deli_df["bowling_team"]=deli_df["bowling_team"].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
deli_df["bowling_team"].unique()
```

```python
deli_df["batting_team"].unique()
```

```python
match["team1"].unique()
```

```python
deli_df['balls_left']=126-(deli_df["over"]*6+deli_df['ball'])
```

```python
deli_df['player_dismissed'] = deli_df['player_dismissed'].fillna("0")
deli_df['player_dismissed'] = deli_df['player_dismissed'].apply(lambda x:x if x == "0" else "1")
deli_df['player_dismissed'] = deli_df['player_dismissed'].astype('int')
wickets = deli_df.groupby('match_id').cumsum()['player_dismissed'].values
deli_df['wickets'] = 10 - wickets
deli_df.head()
```

```python
deli_df.tail()
```

```python
# crr = runs/overs
deli_df['crr'] = (deli_df['current_score']*6)/(120 - deli_df['balls_left'])
```

```python
# rrr
deli_df['rrr'] = (deli_df['runs_left']*6)/deli_df['balls_left']
```

```python
def result(row):
  return 1 if row["batting_team"] == row["winner"] else 0
```

```python
deli_df["result"]=deli_df.apply(result,axis=1)
```

```python
deli_df.head()
```

```python
final_df=deli_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets','total_runs_x','crr','rrr','result']]
```

```python
final_df=final_df.sample(final_df.shape[0])
```

```python
final_df.sample(1)
```

```python
final_df.isnull().sum()
```

```python
final_df.dropna(inplace=True)
```

```python
final_df = final_df[final_df['balls_left'] != 0]
```

```python
final_df.head(2)
```

```python
x=final_df.iloc[:,:-1]
y=final_df.iloc[:,-1]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
trf=ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
],remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
